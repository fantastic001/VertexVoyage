```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
sigmoid = 1 / (1 + np.exp(-x))
plt.plot(x, sigmoid)
plt.title("Sigmoid function")
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.grid()
plt.show() 


# Now, lets see what happens on sigmoid(wx + b) when w and b are random variables. We want to see how the distribution of sigmoid(wx + b) looks like when w and b are random variables. This is important because in node2vec, we have sigmoid(w * (embedding of node i) * (embedding of node j) + b) where w and b are learned parameters. We want to see how the distribution of sigmoid(wx + b) looks like when w and b are random variables.

for w in [1]:
    for b in [-1, 0, 1]:
        x = np.linspace(-10, 10, 100)
        sigmoid = 1 / (1 + np.exp(-w * x - b))
        plt.plot(x, sigmoid, label=f"w={w}, b={b}")
plt.title("Sigmoid function with different w and b")
plt.xlabel("x")
plt.ylabel("sigmoid(wx + b)")
plt.grid()
plt.legend()
plt.show()

# We want to increase probability of connected nodes when dot product approaches mu. We decrease probability of connected nodes when dot product is far from mu. 

x = np.linspace(-20, 20, 100)
mu = 5
sigmoid = 1 / (1 + np.exp(((np.abs(x-mu) - mu))))
plt.plot(x, sigmoid)
plt.title("Sigmoid function with different mu")
plt.xlabel("x")
plt.ylabel("sigmoid(x - mu)")
plt.grid()
plt.show()
```
